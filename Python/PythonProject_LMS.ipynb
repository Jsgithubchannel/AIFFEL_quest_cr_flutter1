# [파이썬] 프로젝트2 확장하기

기준: (1) 메뉴와 가격 리스트를 다차원 리스트로 변경하기
      (2) 메뉴 출력 enumerate() 함수 활용하기
      (3) datetime 모듈로 주문표에 주문 일시 추가하기
      (4) 데코레이터로 주문표 양식 출력하기

---

```python
import datetime

# 주문표 출력 양식을 위한 데코레이터
def print_table_format(func):
    def wrapper(self, *args, **kwargs):
        # 외곽 구분선 출력
        print('==========================================')
        print('                **주문표**                ')
        print('==========================================')

        # 주문 시간 출력
        print(f'주문 시간: {self.order_time.strftime("%Y-%m-%d %H:%M:%S")}')
        print('------------------------------------------')

        # 주문 항목 출력
        func(self, *args, **kwargs)

        # 구분선 출력
        print('------------------------------------------')
        print(f'합계 금액: {self.price_sum} 원')
        print('==========================================')
        print('          감사합니다! 맛있게 드세요 :)        ')
        print('==========================================')
    return wrapper

class Kiosk:
    def __init__(self, menu):
        self.menu = menu  # 메뉴 목록 (이제 다차원 리스트로 구성됨)
        self.price_sum = 0  # 합계 금액 초기화
        self.order_menu = []  # 주문 리스트
        self.order_price = []  # 가격 리스트
        self.order_time = None  # 주문 시간

    # 메뉴 출력
    def menu_print(self):
        print("메뉴:")
        for idx, (item, price) in enumerate(self.menu, 1):
            print(f'{idx}. {item} - {price}원')

    def menu_select(self):
        self.order_menu = []  # 주문 리스트 초기화
        self.order_price = []  # 가격 리스트 초기화

        n = 0
        while n < 1 or len(self.menu) < n:
            n = int(input("음료 번호를 입력하세요 : "))  # 음료 번호 입력

            # 메뉴판에 있는 음료 번호일 때
            if 1 <= n <= len(self.menu):
                item, price = self.menu[n-1]
                # 온도 선택
                t = 0
                while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        temp = "HOT"
                    elif t == 2:
                        temp = "ICE"
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")

                self.order_price.append(price)  # 가격 리스트에 추가
                self.price_sum = price  # 합계 금액
                self.order_menu.append(f'{item} ({temp})')  # 주문 리스트에 추가
                print(f'{item} ({temp}) : {price}원')

            # 메뉴판에 없는 번호일 때
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        # 추가 주문 또는 지불
        while n != 0:  # 지불을 선택하기 전까지 반복합니다.
            print()  # 줄 바꾸면서
            n = int(input("추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))  # 추가 주문 또는 지불
            if n > 0 and n <= len(self.menu):
                item, price = self.menu[n-1]
                # 온도 선택
                t = 0
                while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        temp = "HOT"
                    elif t == 2:
                        temp = "ICE"
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")

                self.order_price.append(price)  # 가격 리스트에 추가
                self.price_sum += price  # 합계 금액 업데이트
                self.order_menu.append(f'{item} ({temp})')  # 주문 리스트에 추가
                print(f'추가 주문 {item} ({temp}) : {price}원\n합계 : {self.price_sum}원')

            else:
                if n == 0:  # 지불을 입력하면
                    print("주문이 완료되었습니다.")
                else:  # 없는 번호를 입력할 때
                    print("없는 메뉴입니다. 다시 주문해 주세요.")

        # 주문 시간이 기록됨
        self.order_time = datetime.datetime.now()

    def pay(self):
        print(f'총 지불할 금액은 {self.price_sum}원 입니다.')

        # 지불 방법 선택
        self.pay_method = 0
        while self.pay_method != 1 and self.pay_method != 2:
            self.pay_method = int(input('현금결제는 1번, 카드결제는 2번을 선택해주세요. : '))
            if self.pay_method == 1:
                print('직원을 호출하겠습니다.')
            elif self.pay_method == 2:
                print('IC칩 방향에 맞게 카드를 꽂아주세요.')
            else:
                print('다시 결제를 시도해 주세요.')

    @print_table_format  # 주문표 출력에 데코레이터 적용
    def table(self):
        # 주문 항목 출력
        for i in range(len(self.order_menu)):
            print(f'{self.order_menu[i]} - {self.order_price[i]}원')

# 전체 메뉴와 가격 (이제 온도는 사용자가 선택)
menu = [
    ['americano', 2000],
    ['latte', 3000],
    ['mocha', 3000],
    ['yuza_tea', 2500],
    ['green_tea', 2500],
    ['choco_latte', 3000]
]

# Kiosk 객체 생성
a = Kiosk(menu)  # menu를 전달하여 객체 생성
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```

# [파이썬] 프로젝트3: 할 일 관리 프로그램 만들기

기준: 다음의 기능을 포함해서 만들기
1) 할 일 목록 보기
2) 할 일 추가
3) 할 일 삭제
4) 할 일 완료 처리

---

```python
class ToDoList:
    def __init__(self):
        self.todo_list = []  # 할 일 목록 초기화

    def display_tasks(self):
        """할 일 목록 보기"""
        if not self.todo_list:
            print("할 일이 없습니다!")
        else:
            print("현재 할 일 목록:")
            for index, task in enumerate(self.todo_list, 1):
                status = "완료" if task['completed'] else "진행 중"
                print(f"{index}. {task['task']} - {status}")

    def add_task(self, task_name):
        """할 일 추가"""
        task = {"task": task_name, "completed": False}  # 할 일과 완료 상태를 딕셔너리로 저장
        self.todo_list.append(task)
        print(f"할 일 '{task_name}'이(가) 추가되었습니다.")

    def delete_task(self, task_index):
        """할 일 삭제"""
        if 1 <= task_index <= len(self.todo_list):
            removed_task = self.todo_list.pop(task_index - 1)
            print(f"할 일 '{removed_task['task']}'이(가) 삭제되었습니다.")
        else:
            print("잘못된 번호입니다. 다시 입력해 주세요.")

    def mark_completed(self, task_index):
        """할 일 완료 처리"""
        if 1 <= task_index <= len(self.todo_list):
            task = self.todo_list[task_index - 1]
            if task['completed']:
                print(f"할 일 '{task['task']}'은(는) 이미 완료된 상태입니다.")
            else:
                task['completed'] = True
                print(f"할 일 '{task['task']}'이(가) 완료 처리되었습니다.")
        else:
            print("잘못된 번호입니다. 다시 입력해 주세요.")

    def menu(self):
        """메뉴 출력"""
        print("\n할 일 관리 프로그램")
        print("1. 할 일 목록 보기")
        print("2. 할 일 추가")
        print("3. 할 일 삭제")
        print("4. 할 일 완료 처리")
        print("0. 종료")

    def run(self):
        """프로그램 실행"""
        while True:
            self.menu()  # 메뉴 출력
            choice = input("선택하세요: ")

            if choice == "1":
                self.display_tasks()
            elif choice == "2":
                task_name = input("추가할 할 일을 입력하세요: ")
                self.add_task(task_name)
            elif choice == "3":
                self.display_tasks()
                task_index = int(input("삭제할 할 일 번호를 입력하세요: "))
                self.delete_task(task_index)
            elif choice == "4":
                self.display_tasks()
                task_index = int(input("완료 처리할 할 일 번호를 입력하세요: "))
                self.mark_completed(task_index)
            elif choice == "0":
                print("프로그램을 종료합니다.")
                break
            else:
                print("잘못된 선택입니다. 다시 입력해 주세요.")

# 실행 예시
todo_app = ToDoList()
todo_app.run()
```

```python
import datetime

# 주문표 출력 양식을 위한 데코레이터
def print_table_format(func):
    def wrapper(self, *args, **kwargs):
        # 외곽 구분선 출력
        print('==========================================')
        print('          **여니왕 타코 주문표**          ')
        print('==========================================')

        # 주문 시간 출력
        print(f'주문 시간: {self.order_time.strftime("%Y-%m-%d %H:%M:%S")}')
        print('------------------------------------------')

        # 주문 항목 출력
        func(self, *args, **kwargs)

        # 구분선 출력
        print('------------------------------------------')
        print(f'합계 금액: {self.price_sum} 원')
        print('==========================================')
        print('          감사합니다! 맛있게 드세요 :)        ')
        print('==========================================')
    return wrapper

class Kiosk:
    def __init__(self, quantity, flavors, toppings, drinks):
        self.quantity = quantity  # 수량 리스트
        self.flavors = flavors  # 맛 리스트
        self.toppings = toppings  # 토핑 리스트
        self.drinks = drinks  # 음료 리스트
        self.price_sum = 0  # 합계 금액 초기화
        self.order_details = []  # 주문 내용
        self.order_time = None  # 주문 시간

    # 수량 선택
    def select_quantity(self):
        print("수량을 선택하세요:")
        for idx, (quantity, price) in enumerate(self.quantity, 1):
            print(f"{idx}. {quantity}알 (+{price}원)")

        choice = int(input("수량을 선택하세요 (1-7): "))
        quantity, price = self.quantity[choice - 1]
        self.price_sum += price
        self.order_details.append(f"수량: {quantity}알 (+{price}원)")
        print(f"{quantity}알이 선택되었습니다. 가격은 +{price}원입니다.")

    # 맛 선택
    def select_flavor(self):
        print("\n맛을 선택하세요:")
        for idx, (flavor, extra_price) in enumerate(self.flavors, 1):
            print(f"{idx}. {flavor} (+{extra_price}원)" if extra_price > 0 else f"{idx}. {flavor} (+0원)")

        choice = int(input("맛을 선택하세요 (1-16): "))
        flavor, extra_price = self.flavors[choice - 1]
        self.price_sum += extra_price
        self.order_details.append(f"맛: {flavor} (+{extra_price}원)")
        print(f"'{flavor}' 맛이 선택되었습니다. 가격은 +{extra_price}원입니다.")

    # 토핑 선택
    def select_topping(self):
        print("\n토핑을 선택하세요 (무료):")
        for idx, topping in enumerate(self.toppings, 1):
            print(f"{idx}. {topping} (무료)")

        topping_choice = int(input("토핑을 선택하세요 (1-3): "))
        topping = self.toppings[topping_choice - 1]
        self.order_details.append(f"토핑: {topping} (무료)")
        print(f"'{topping}' 토핑이 선택되었습니다.")

    # 음료 선택 (여러 개 선택 가능)
    def select_drink(self):
        while True:
            print("\n음료를 선택하세요:")
            for idx, (drink, price) in enumerate(self.drinks, 1):
                print(f"{idx}. {drink} (+{price}원)")

            drink_choice = int(input("음료를 선택하세요 (1-9): "))
            drink, price = self.drinks[drink_choice - 1]
            self.price_sum += price
            self.order_details.append(f"음료: {drink} (+{price}원)")
            print(f"'{drink}' 음료가 선택되었습니다. 가격은 +{price}원입니다.")

            add_more = input("음료를 더 추가하시겠습니까? (y/n): ")
            if add_more.lower() != 'y':
                break


    # 주문 시간 기록
    def record_order_time(self):
        self.order_time = datetime.datetime.now()

    # 총합 계산 및 결제
    def pay(self):
        print(f'총 지불할 금액은 {self.price_sum}원 입니다.')

        # 지불 방법 선택
        self.pay_method = 0
        while self.pay_method != 1 and self.pay_method != 2:
            self.pay_method = int(input('현금결제는 1번, 카드결제는 2번을 선택해주세요. : '))
            if self.pay_method == 1:
                print('직원을 호출하겠습니다.')
            elif self.pay_method == 2:
                print('IC칩 방향에 맞게 카드를 꽂아주세요.')
            else:
                print('다시 결제를 시도해 주세요.')

    @print_table_format  # 주문표 출력에 데코레이터 적용
    def table(self):
        # 주문 항목 출력
        for detail in self.order_details:
            print(detail)

# 각 항목에 대한 리스트
quantity = [
    ('6', 5100),
    ('9', 7200),
    ('12', 9200),
    ('16', 11900),
    ('18', 13200),
    ('24', 17600),
    ('27', 19600)
]

flavors = [
    ('여니왕시그니처', 300),
    ('청양마요', 300),
    ('핵매운맛', 300),
    ('데리야끼', 300),
    ('데리야끼(매콤)', 0),
    ('갈릭치즈', 300),
    ('갈릭치즈(매콤)', 300),
    ('나쵸체다치즈', 300),
    ('나쵸체다치즈(매콤)', 300),
    ('불닭마요', 300),
    ('치즈불닭마요', 300),
    ('오사카정통소스', 300),
    ('짭짤한 명란마요', 500),
    ('짭짤한 명란마요(매콤)', 500),
    ('고소한 명란마요', 500),
    ('고소한 명란마요(매콤)', 500)
]

toppings = [
    '대파토핑', '청양고추토핑', '와사비토핑'
]

drinks = [
    ('코카콜라 500ml', 2000),
    ('제로콜라 500ml', 2000),
    ('복숭아아이스티', 3000),
    ('체리콕콕', 3500),
    ('청포도 에이드', 3500),
    ('자몽 에이드', 3500),
    ('레몬 에이드', 3500),
    ('식혜', 3000),
    ('파인애플 환타 500ml', 2000),
    ('스프라이트 500ml', 2000)
]

# Kiosk 객체 생성
kiosk = Kiosk(quantity, flavors, toppings, drinks)

# 주문 과정
kiosk.select_quantity()
kiosk.select_flavor()
kiosk.select_topping()
kiosk.select_drink()
kiosk.record_order_time()

# 결제
kiosk.pay()

# 주문표 출력
kiosk.table()
```
