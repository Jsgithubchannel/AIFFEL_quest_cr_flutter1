```python
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')

        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')

        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    def par(self):
        par_width, height = map(int, input('넓이, 높이를 입력하세요. 예시: 넓이,높이\n >>>').split(','))
        area = par_width * height
        result = '평행사변형의 넓이는 : ' + str(area)
        return result

    def trape(self):
        top_width, bot_width, height = map(int, input('윗변, 아랫변, 높이를 입력하세요. 예시: 윗변,아랫변,높이\n >>>').split(','))
        area = (top_width + bot_width) * height * 0.5
        result = '사다리꼴의 넓이는 : ' + str(area)
        return result

a = Square()
#a.rect()
#a.par()
#a.trape()
```

# [파이썬] 프로젝트2 확장하기

기준: (1) 메뉴와 가격 리스트를 다차원 리스트로 변경하기
      (2) 메뉴 출력 enumerate() 함수 활용하기
      (3) datetime 모듈로 주문표에 주문 일시 추가하기
      (4) 데코레이터로 주문표 양식 출력하기

---

```python
import datetime

# 주문표 출력 양식을 위한 데코레이터
def print_table_format(func):
    def wrapper(self, *args, **kwargs):
        # 외곽 구분선 출력
        print('==========================================')
        print('                **주문표**                ')
        print('==========================================')

        # 주문 시간 출력
        print(f'주문 시간: {self.order_time.strftime("%Y-%m-%d %H:%M:%S")}')
        print('------------------------------------------')

        # 주문 항목 출력
        func(self, *args, **kwargs)

        # 구분선 출력
        print('------------------------------------------')
        print(f'합계 금액: {self.price_sum} 원')
        print('==========================================')
        print('          감사합니다! 맛있게 드세요 :)        ')
        print('==========================================')
    return wrapper

class Kiosk:
    def __init__(self, menu):
        self.menu = menu  # 메뉴 목록 (이제 다차원 리스트로 구성됨)
        self.price_sum = 0  # 합계 금액 초기화
        self.order_menu = []  # 주문 리스트
        self.order_price = []  # 가격 리스트
        self.order_time = None  # 주문 시간

    # 메뉴 출력
    def menu_print(self):
        print("메뉴:")
        for idx, (item, price) in enumerate(self.menu, 1):
            print(f'{idx}. {item} - {price}원')

    def menu_select(self):
        self.order_menu = []  # 주문 리스트 초기화
        self.order_price = []  # 가격 리스트 초기화

        n = 0
        while n < 1 or len(self.menu) < n:
            n = int(input("음료 번호를 입력하세요 : "))  # 음료 번호 입력

            # 메뉴판에 있는 음료 번호일 때
            if 1 <= n <= len(self.menu):
                item, price = self.menu[n-1]
                # 온도 선택
                t = 0
                while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        temp = "HOT"
                    elif t == 2:
                        temp = "ICE"
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")

                self.order_price.append(price)  # 가격 리스트에 추가
                self.price_sum = price  # 합계 금액
                self.order_menu.append(f'{item} ({temp})')  # 주문 리스트에 추가
                print(f'{item} ({temp}) : {price}원')

            # 메뉴판에 없는 번호일 때
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        # 추가 주문 또는 지불
        while n != 0:  # 지불을 선택하기 전까지 반복합니다.
            print()  # 줄 바꾸면서
            n = int(input("추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))  # 추가 주문 또는 지불
            if n > 0 and n <= len(self.menu):
                item, price = self.menu[n-1]
                # 온도 선택
                t = 0
                while t != 1 and t != 2:  # 1이나 2를 입력할 때까지 물어보기
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        temp = "HOT"
                    elif t == 2:
                        temp = "ICE"
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")

                self.order_price.append(price)  # 가격 리스트에 추가
                self.price_sum += price  # 합계 금액 업데이트
                self.order_menu.append(f'{item} ({temp})')  # 주문 리스트에 추가
                print(f'추가 주문 {item} ({temp}) : {price}원\n합계 : {self.price_sum}원')

            else:
                if n == 0:  # 지불을 입력하면
                    print("주문이 완료되었습니다.")
                else:  # 없는 번호를 입력할 때
                    print("없는 메뉴입니다. 다시 주문해 주세요.")

        # 주문 시간이 기록됨
        self.order_time = datetime.datetime.now()

    def pay(self):
        print(f'총 지불할 금액은 {self.price_sum}원 입니다.')

        # 지불 방법 선택
        self.pay_method = 0
        while self.pay_method != 1 and self.pay_method != 2:
            self.pay_method = int(input('현금결제는 1번, 카드결제는 2번을 선택해주세요. : '))
            if self.pay_method == 1:
                print('직원을 호출하겠습니다.')
            elif self.pay_method == 2:
                print('IC칩 방향에 맞게 카드를 꽂아주세요.')
            else:
                print('다시 결제를 시도해 주세요.')

    @print_table_format  # 주문표 출력에 데코레이터 적용
    def table(self):
        # 주문 항목 출력
        for i in range(len(self.order_menu)):
            print(f'{self.order_menu[i]} - {self.order_price[i]}원')

# 전체 메뉴와 가격 (이제 온도는 사용자가 선택)
menu = [
    ['americano', 2000],
    ['latte', 3000],
    ['mocha', 3000],
    ['yuza_tea', 2500],
    ['green_tea', 2500],
    ['choco_latte', 3000]
]

# Kiosk 객체 생성
a = Kiosk(menu)  # menu를 전달하여 객체 생성
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```
