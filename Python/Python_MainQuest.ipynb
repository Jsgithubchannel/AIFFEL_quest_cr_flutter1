```python
#1 Account 클래스 생성
#2 클래스 변수: 생성된 계좌 객체의 개수 저장
#3 클래스 변수 출력
#4 입금 메서드
#5 출금 메서드
#6 정보 출력 메서드
#7 이자 지급하기
#8 여러 객체 생성
#9 잔고가 100만원 이상인 고객의 정보만 출력
#10 입금과 출금 내역 출력
#11 입금과 출금 내역 테스트

import random

#1 Account 클래스 생성
class Account:

    #2 클래스 변수: 생성된 계좌 객체의 개수 저장
    account_count = 0

    # 생성자에서는 예금주와 초기 잔액만 입력받음
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

        # 은행이름은 SC은행으로 계좌번호는 3자리-2자리-6자리 형태
        self.bank_name = "SC은행"
        self.account_number = f"{random.randint(100, 999)}-{random.randint(10, 99)}-{random.randint(100000, 999999)}"

        # 계좌 객체가 생성될 때마다 계좌 개수 증가
        Account.account_count += 1

        # 입출금 관련 변수 초기화
        self.deposit_count = 0
        self.deposit_history_list = []
        self.withdraw_history_list = []
        self.interest_history_list = []  # 이자 지급 내역 리스트 추가

    #3 클래스 변수 출력
    def get_account_num(self):
        return Account.account_count

    #4 입금 메서드
    def deposit(self, amount):
        if amount < 1:
            print("입금은 최소 1원 이상만 가능합니다.")
        else:
            self.balance += amount
            self.deposit_count += 1
            # 입금 내역 기록
            self.deposit_history_list.append(f"{self.deposit_count}회: 입금 금액: {amount:,} 잔액: {self.balance:,}")

            #7 이자 지급하기: 입금 횟수가 5회가 될 때
            if self.deposit_count % 5 == 0:
                interest = self.balance * 0.01  # 1% 이자
                self.balance += interest
                self.interest_history_list.append(f"입금 {self.deposit_count}회차 이자: {interest:,.0f}원")
                print(f"이자 지급! {interest:,.0f}원 이자가 입금되었습니다.")

            # 입금 후 바로 내역 출력
            print(f"{self.owner}님의 입금 내역:")
            for entry in self.deposit_history_list:
                print(entry)

    #5 출금 메서드
    def withdraw(self, amount):
        if amount > self.balance:
            print("출금은 계좌의 잔고 이상으로 출금할 수는 없습니다.")
        elif amount < 1:
            print("출금 금액은 1원 이상이어야 합니다.")
        else:
            self.balance -= amount
            self.withdraw_history_list.append(f"출금 금액: {amount:,} 잔액: {self.balance:,}")
            print(f"{self.owner}님, 출금 내역이 기록되었습니다. 현재 잔액은 {self.balance:,}원입니다.")

    #6 정보 출력 메서드
    def display_info(self):
        print(f"은행이름: {self.bank_name}, 예금주: {self.owner}, 계좌번호: {self.account_number}, 잔액: {self.balance:,}원")

    #10 입출금 내역 출력
    def deposit_history(self):
        print("입금 내역:")
        for entry in self.deposit_history_list:
            print(entry)

    def withdraw_history(self):
        print("출금 내역:")
        for entry in self.withdraw_history_list:
            print(entry)

    def interest_history(self):
        if self.interest_history_list:
            print("이자 내역:")
            for entry in self.interest_history_list:
                print(entry)
        else:
            print("이자 지급 내역이 없습니다.")

#8 여러 객체 생성
accounts = [
    Account("곽현정", 10000000),
    Account("황유진", 5000000),
    Account("정승연", 3000000),
    Account("황인찬", 400000)
]

#9 잔고가 100만원 이상인 고객의 정보만 출력
print("잔고가 100만원 이상인 고객 정보:")
for account in accounts:
    if account.balance >= 1000000:
        account.display_info()

#10 입출금 내역을 사용자에게 입력받기
for account in accounts:
    print(f"\n{account.owner}님의 계좌에서 입금하거나 출금할 금액을 입력하세요.")
    deposit_amount = int(input(f"{account.owner}님, 입금할 금액을 입력하세요: "))
    account.deposit(deposit_amount)

    withdraw_amount = int(input(f"{account.owner}님, 출금할 금액을 입력하세요: "))
    account.withdraw(withdraw_amount)

#10 입출금 및 이자 내역 출력
for account in accounts:
    print(f"\n{account.owner}님의 입출금 내역:")
    account.deposit_history()
    account.withdraw_history()
    account.interest_history()3
```

: battle 함수

:몬스터의 이름, 레벨이 매핑된 딕셔너리 정의하기

*monster_dict = {‘슬라임’: 1, ‘고블린’: 2, ‘오크’: 3}
사용자로부터 이름을 입력받아 Player 인스턴스 생성하기

*몬스터 딕셔너리로부터 Monster 인스턴스 생성하기

*생성된 Monster 인스턴스와 Player 인스턴스가 battle 함수를 통해 전투

*Player는 생성된 몬스터 3마리(슬라임, 고블린, 오크)와 모두 전투해야함

*전투 도중에 Player가 사망하면 이후 전투를 진행하지 않고 ‘게임오버’ 출력하고 종료




```python
import random
import time

# 조사 처리 함수
def get_correct_particle(name, particle_with, particle_without):
    """
    이름과 함께 적절한 조사를 반환하는 함수.
    받침이 있으면 particle_with, 없으면 particle_without를 반환.
    """
    last_char = name[-1]
    # 이름의 마지막 문자가 받침이 있는지 없는지 확인
    if (ord(last_char) - 44032) % 28 == 0:  # 받침이 없는 경우
        return particle_without  # 받침 없으면 "이"
    return particle_with  # 받침 있으면 "가"

# Q1. Character 클래스 만들기
class Character:
    def __init__(self, name, level, health, attack, defense):
        self.name = name
        self.level = level
        self.health = health
        self.attack = attack
        self.defense = defense

    def is_alive(self):
        return self.health > 0

    def take_damage(self, damage):
        actual_damage = max(0, damage - self.defense)
        self.health -= actual_damage
        return actual_damage

    def attack_target(self, target):
        damage = random.randint(1, self.attack)
        actual_damage = target.take_damage(damage)
        return actual_damage

# Q2-1. Player 클래스 만들기
class Player(Character):
    def __init__(self, name):
        super().__init__(name, level=1, health=100, attack=25, defense=5)
        self.experience = 0

    def gain_experience(self, amount):
        self.experience += amount
        self.level_up()

    def level_up(self):
        while self.experience >= 50:
            self.experience -= 50
            self.level += 1
            self.attack += 10
            self.defense += 5
            particle = get_correct_particle(self.name, "가", "이")  # "가/이" 조사 처리
            # 레벨업 메시지 수정
            print(f"{self.name}{particle} 레벨업 했습니다! 레벨: {self.level}, 공격력: {self.attack}, 방어력: {self.defense}")
            time.sleep(1)

# Q2-2. Monster 클래스 만들기
class Monster(Character):
    def __init__(self, name, level):
        health = random.randint(10, 30) * level
        attack = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name, level, health, attack, defense)

# Q3. battle 함수 만들기
def battle(player, monster):
    # "과/와" 조사 처리
    particle = get_correct_particle(monster.name, "과", "와")
    print(f"{monster.name}{particle}의 전투를 시작합니다!\n")
    time.sleep(1)

    while player.is_alive() and monster.is_alive():
        # Player 공격
        damage = player.attack_target(monster)
        print(f"{player.name}{get_correct_particle(player.name, '이', '가')} {monster.name}에게 {damage}만큼 공격했다...! {monster.name}의 체력: {monster.health}")
        time.sleep(1)

        if not monster.is_alive():
            print(f"{monster.name}을(를) 이겼다!")
            player.gain_experience(monster.level * 20)
            print("전투 승리!\n")
            time.sleep(1)
            return

        # Monster 공격
        damage = monster.attack_target(player)
        print(f"{monster.name}{get_correct_particle(monster.name, '이', '가')} {player.name}에게 {damage}만큼 공격했다...! {player.name}의 체력: {player.health}")
        time.sleep(1)

        if not player.is_alive():
            print("전투 패배..")
            return

# Q4. main 함수 만들기
def main():
    player_name = input("플레이어 이름을 입력하세요: ")
    player = Player(player_name)

    monster_dict = {'슬라임': 1, '고블린': 2, '오크': 3}

    # 3마리 몬스터와 전투
    for monster_name, level in monster_dict.items():
        monster = Monster(monster_name, level)
        battle(player, monster)
        if not player.is_alive():
            print("게임오버")
            break

# 게임 실행
if __name__ == "__main__":
    main()
```
